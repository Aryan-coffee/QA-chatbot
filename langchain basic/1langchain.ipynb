```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
```

building your first chain


```python
def create_story_chain():
    story_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", "You are a creative storyteller. Write a short and engaging story based on a given theme"),
            ("user", "Theme: {theme}\nMain character: {character}\nSetting: {setting}\nLength: {length}")
        ]
    )
    
    analysis_prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a story analyst. Analyze the given story and provide insights on its plot, characters, and themes."),
        ("user", "{story}")
    ])

    story_chain = (
        story_prompt | model | StrOutputParser()
    )

    def analyze_story(story_text):
        return {"story": story_text}

    analysis_chain = (
        story_chain
        | RunnableLambda(analyze_story)
        | analysis_prompt
        | model
        | StrOutputParser()
    )

    return analysis_chain
```

```python
chain=create_story_chain()
chain
```

```python
result = chain.invoke({
    "theme": "Artificial Intelligence",
    "character": "A curious robot named Robo",
    "setting": "A futuristic city where robots and humans coexist",
})

print("Story and Analysis:")
print(result)
```

```python
def create_story_chain():
    story_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", "You are a creative storyteller. Write a short and engaging story based on a given theme."),
            ("user", "Theme: {theme}\nMain character: {character}\nSetting: {setting}\nLength: {length}")
        ]
    )

    analysis_prompt = ChatPromptTemplate.from_messages(
        [
            ("system", "You are a story analyst. Analyze the given story and provide insights on its plot, characters, and themes."),
            ("user", "{story}")
        ]
    )

    story_chain = story_prompt | model | StrOutputParser()
    analysis_chain = analysis_prompt | model | StrOutputParser()

    return story_chain, analysis_chain

story_chain, analysis_chain = create_story_chain()

story = story_chain.invoke({
    "theme": "Artificial Intelligence",
    "character": "A curious robot named Robo",
    "setting": "A futuristic city where robots and humans coexist",
    "length": "Short",
})

analysis = analysis_chain.invoke({"story": story})

print("Story:")
print(story)
print("\nAnalysis:")
print(analysis)
```
